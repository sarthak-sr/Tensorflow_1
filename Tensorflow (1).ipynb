- The tf.concat() function in TensorFlow is used to concatenate or join tensors along a specified axis. It allows you to combine multiple tensors into a single tensor. Here are a few conditions to keep in mind when using tf.concat():

- The shapes of the tensors being concatenated must be the same in all dimensions except the concatenation axis. This means all dimensions, except the concatenation axis, should have the same size in each tensor. For example, if you are concatenating tensors along axis 0, the sizes of the other axes for all tensors should match.

- The data types of the tensors being concatenated should be compatible. TensorFlow will raise an error if the data types are not compatible, such as trying to concatenate tensors with different data types, like combining float32 and int32 tensors.

- The concatenation axis should be a valid axis for the given tensors. The axis number should fall within the range of the tensor's dimensions. For example, if you have a 3-dimensional tensor, you can choose any axis from 0 to 2 to concatenate along.

```python
t1 = tf.constant([[1, 2, 3], [4, 5, 6]])
t2 = tf.constant([[7, 8, 9,13], [10, 11, 12,14]])
print(t1.shape)
print(t2.shape)
print(tf.concat([t1,t2], 1))
```

28. tf.stack()

29.   tf.pad(
    tensor, paddings, mode='CONSTANT', constant_values=0, name=None
)

- This operation pads a tensor according to the paddings you specify. paddings is an integer tensor with shape [n, 2], where n is the rank of tensor. For each dimension D of input, paddings[D, 0] indicates how many values to add before the contents of tensor in that dimension, and paddings[D, 1] indicates how many values to add after the contents of tensor in that dimension. If mode is "REFLECT" then both paddings[D, 0] and paddings[D, 1] must be no greater than tensor.dim_size(D) - 1. If mode is "SYMMETRIC" then both paddings[D, 0] and paddings[D, 1] must be no greater than tensor.dim_size(D).

- The tf.pad() function in TensorFlow is used to pad a tensor with values. It allows you to add values to the boundaries of a tensor, effectively increasing its size in one or more dimensions. Here are a few conditions to keep in mind when using tf.pad():

- The input tensor should have a known shape. TensorFlow requires the shape of the input tensor to be known during graph construction. If the shape is dynamic or unknown, you may need to use other methods, like tf.pad() within a tf.function or tf.py_function to handle dynamic shapes.

- The paddings argument should be a tensor or a list of pairs. The paddings argument specifies how the tensor should be padded in each dimension. It should be of shape (n, 2), where n is the number of dimensions of the input tensor. Each row of paddings represents the padding configuration for a particular dimension, with the first value indicating the padding before the dimension and the second value indicating the padding after the dimension.

- The padding values should be compatible with the data type of the input tensor. The padding values you provide should have the same data type as the input tensor. TensorFlow will raise an error if the data types are not compatible.

Example:
**t1 is 2D tensor. Therefore it has two dimestions or axis. The padding variable as to be a list of lists or a tensor. The first element of the represents the first dimention i.e. along the rows, now if we want to pad above as well as below, we can write something as [1,1], this will pad 1 row above and 1 row below. if we write some this as [2,1], this is will add 2 rows above and only one row below. The next element of the padding list or tensor will be seccond dimention i.e. along the columns so [1,1] will add one column on each side, [2,1] will add 2 columns on the LEFT and one on the RIGHT.**

```python
print(t1)

padding = tf.constant([[1,1],[0,0]])
print(tf.pad(t1, padding, mode='CONSTANT', constant_values=0, name=None))
print()
padding = tf.constant([[2,1],[0,0]])
print(tf.pad(t1, padding, mode='CONSTANT', constant_values=0, name=None))
print()
padding = tf.constant([[0,0],[1,1]])
print(tf.pad(t1, padding, mode='CONSTANT', constant_values=0, name=None))
print()
padding = tf.constant([[0,0],[2,1]])
print(tf.pad(t1, padding, mode='CONSTANT', constant_values=0, name=None))
print()
padding = tf.constant([[1,1],[2,2]])
print(tf.pad(t1, padding, mode='CONSTANT', constant_values=0, name=None))
```

30. tf.gather():

tf.gather() is a TensorFlow function that allows you to gather specific elements from a tensor along a specified axis. It takes two main arguments: params and indices. Here's an explanation of each argument and an example to demonstrate how tf.gather() works:

- params: This argument represents the tensor from which you want to gather elements. It can be a 1-D, 2-D, or higher-dimensional tensor.

- indices: This argument specifies the indices of the elements you want to gather from params. It can be an integer tensor or a scalar.

The shape of the output tensor produced by tf.gather() depends on the shape of the indices tensor. The dimensions of the output tensor are the same as the indices tensor, and the values at each index are gathered from the corresponding indices in the params tensor.

**Indexing will either give you continuos or a single value, i.e. either it will give you the first row OR the first to the third row. But GATHERING can get you the exact rows example first and third, no seccond.**

```python
print(t1)

print(tf.gather(t1, [0,1,0,1])) #By default the axis is 0
```

```python
print(tf.gather(t1, [0,1,0,1], axis = 1))
```

31. tf.gather_nd()

- tf.gather(): This function allows you to gather elements from a tensor params based on the indices provided in the indices tensor. The shape of the output tensor depends on the shape of the indices tensor.

- tf.gather_nd(): This function is similar to tf.gather(), but it allows you to gather elements from a tensor params based on multi-dimensional indices provided in the indices tensor. The shape of the output tensor is determined by the shape of the indices tensor.

```python
params = tf.constant([[1, 2], [3, 4], [5, 6]])  # 2-D tensor
indices = tf.constant([[0, 0], [2, 1]])  # Gather elements at (0, 0) and (2, 1) positions

gathered_values = tf.gather_nd(params, indices)

print(gathered_values)

indices = tf.constant([[0, 0], [2, 1]])

gathered_values = tf.gather_nd(params, indices)

print(gathered_values)
```

37. Variable:

https://www.tensorflow.org/api_docs/python/tf/Variable

The tf.Variable() is an important concept in TensorFlow as it allows you to create and manage trainable variables within your machine learning models. These variables hold the state of the model and can be updated during training.

Here are a few reasons why tf.Variable() is important:

- Trainable Parameters: tf.Variable() is used to create variables that hold the learnable parameters of your model. During training, these variables are updated by optimization algorithms (e.g., gradient descent) to minimize the loss function and improve model performance.

**If trainable parameter while defining a variable is set to False it wont update upon training the model.**

- Persistent State: Variables created with tf.Variable() persist their values across multiple executions of your model. This allows you to save and restore the trained model parameters, making it easier to deploy and use the model in production.

- Automatic Differentiation: TensorFlow's automatic differentiation engine keeps track of the operations involving variables, allowing you to compute gradients for backpropagation automatically. This is crucial for training neural networks and updating the variables' values based on the computed gradients.

Some of the most commonly used methods associated with tf.Variable() include:

1. assign(): This method allows you to update the value of a variable. For example, you can use variable.assign(new_value) to assign a new value to the variable.

2. assign_add() and assign_sub(): These methods let you increment or decrement the value of a variable by a specified amount. For example, variable.assign_add(1) increases the value of the variable by 1.

3. read_value(): This method returns the current value of the variable as a TensorFlow tensor. You can use variable.read_value() to access the value of the variable.

4. initializer(): This method returns an operation that initializes the variable. You can use variable.initializer to get the initializer operation associated with the variable.

Example:


```python
x = tf.constant([1,2])
x_var = tf.Variable(x, name = 'var1')
print(x_var)
x_var.assign_sub([3,4])
print(x_var)
```

- The tf.device() function is used in TensorFlow to specify the device (e.g., CPU or GPU) on which a particular computation or operation should be executed. It allows you to control the placement of operations and tensors to utilize specific hardware resources efficiently.

Here's how tf.device() works:

1. Specify Device: You can use tf.device() as a context manager to set the device for a block of code. For example, with tf.device('/GPU:0'): specifies that the following operations should be executed on the first GPU device available.

2. Device Placement: TensorFlow automatically assigns devices to operations based on their compatibility and availability. However, you can use tf.device() to override this automatic placement and explicitly assign operations to specific devices.

3. Fallback Devices: If the specified device is not available, TensorFlow will use a fallback strategy. For example, if you specify /GPU:0 but there are no GPUs available, TensorFlow will automatically fall back to using the CPU.

4. Multi-GPU Systems: If you have multiple GPUs, you can specify a specific GPU by its index (e.g., /GPU:0 for the first GPU, /GPU:1 for the second GPU, and so on).

```python
with tf.device('CPU:0'):
  x_var = tf.Variable(0.2)

print(x_var.device)

#In google collab need to change CPU -> GPU from Runtime -> Change Runtime Type -> T4 GPU.

with tf.device('GPU:0'): #Can be used to define any variable.
  x3 = tf.constant([1,1,1])

print(x3.device)
```
